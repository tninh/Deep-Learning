### 1-1. Fitting the data using linear regression model

```python
df = pd.read_csv("ex2data1.csv")
```

```python
x_train = df.iloc[:,0].values
y_train = df.iloc[:,1].values
```

```python
y_train = np.matrix(y_train)
```

```python
x_train = np.matrix(x_train)
```

```python
plt.scatter([x_train], [y_train], color="red", marker="x")
plt.title("Scatter plot data")
plt.show()
```

```python
x_train = np.insert(x_train, 0, 1, axis = 0)
```

```python
x_train = x_train.T
y_train = y_train.T
```

```python
def initialize_parameters(x):
    theta = np.random.rand(1, x.shape[1])
    theta = np.matrix(theta)
    return theta
```

```python
def h(x, theta):
    y_predict = x * theta.T
    return y_predict
```

```python
def cost_function(h, y):
    m = y.shape[0]
    total_loss = np.square(np.subtract(h, y)).sum(axis=0)
    cost = (1./(2. * m)) * total_loss
    cost = np.asscalar(cost)
    return cost
```

```python
def fit(x, y, num_iterations=100, learning_rate=0.001):
    theta = initialize_parameters(x)
    m = x.shape[0]
    costs = []
    for i in range(num_iterations):
        for j in range(x.shape[1]):
            theta[0, j] = theta[0, j] - learning_rate * ((1.0/m)*((np.multiply(np.subtract(h(x, theta), y),(x[:, j]))).sum(axis = 0)))
        y_predict = h(x, theta)
        costs.append(cost_function(y_predict, y_train))
        if i % 10000 == 0:
            print("Cost after {} iterations: {}".format(i, costs[i]))
    return theta, costs
```

```python
#theta = initialize_parameters(x_train)
theta, costs = fit(x_train, y_train, num_iterations=5000, learning_rate = 0.1)
```

```python
y_predict = h(x_train, theta)
```

- Optimized Weights

```python
print("W0= {}".format(np.asscalar(theta[:,0])))
print("W1= {}".format(np.asscalar(theta[:,1])))
```

```python
plt.scatter([x_train[:,1]],[y_train],40,color="red",marker="x")
plt.plot(x_train[:,1],y_predict)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression best line fitted to training examples")
plt.show()
```

- RMSE

```python
np.sqrt(cost_function(y_predict, y_train))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
x_train[:,1] = scaler.fit_transform(x_train[:,1])
```

```python
def genPolyFeatures(myX,p):
    """
    Function takes in the X matrix (with bias term already included as the first column)
    and returns an X matrix with "p" additional columns.
    The first additional column will be the 2nd column (first non-bias column) squared,
    the next additional column will be the 2nd column cubed, etc.
    """
    newX = myX.copy()
    for i in xrange(p):
        dim = i+2
        newX = np.insert(newX,newX.shape[1],np.power(newX[:,1],dim).flatten(),axis=1)
    return newX
```

### 1-5 Fit data using 16th order regression with ridge 

```python
def ridge_cost_function(h, y, theta, ld = 0.1):
    m = y.shape[0]
    total_loss = np.square(np.subtract(h, y)).sum(axis=0)
    cost = (1./(2. * m)) * total_loss
    reg_term = (ld / m) * np.sum(theta, axis=1)
    cost = cost + reg_term
    cost = np.asscalar(cost)
    return cost
```

```python
def fit_ridge(x, y, num_iterations=1000, learning_rate=0.1, ld = 0.1):
    theta = initialize_parameters(x)
    m = x.shape[0]
    costs = []
    for i in range(num_iterations):
        for j in range(1, x.shape[1]):
            theta[0,0] = theta[0,0] - learning_rate * ((1.0/m)*((np.multiply(np.subtract(h(x, theta), y),(x[:, 0]))).sum(axis = 0)))
            theta[0, j] = theta[0, j] * (1. - learning_rate * ((2. * ld)/m)) - learning_rate * ((1.0/m)*((np.multiply(np.subtract(h(x, theta), y),(x[:, j]))).sum(axis = 0)))
        y_predict = h(x, theta)
        costs.append(ridge_cost_function(y_predict, y_train, theta, ld))
        if i % 10000 == 0:
            print("Cost after {} iterations: {}".format(i, costs[i]))
    return theta, costs
```

```python
theta5, costs5 = fit_ridge(x_train_4, y_train, num_iterations=20000, learning_rate=0.01, ld = 0.001)
```

```python
y_predict_5 = h(x_train_4, theta5)
x_sorted = np.sort(x_train_4, axis=0)
y_predict_sorted = h(x_sorted, theta5)
plt.scatter([x_train[:,1]],[y_train],40,color="red",marker="x")
plt.plot(x_sorted[:,1], y_predict_sorted)
plt.xlabel("x")
plt.ylabel("y")
plt.title("16th order polynomial fitted to training examples with lambda = 0.001")
plt.show()
```

### 3-3 Fit the data using TensorFlow - Ridge regression

```python
import tensorflow as tf
x_train = tf.constant(x_train_1, dtype=tf.float32, name='X')
x_val = tf.constant(x_val_1, dtype=tf.float32, name='X_val')
x_test = tf.constant(x_test_1, dtype=tf.float32, name='X_test')
```

```python
y_train = tf.constant(y_train, dtype=tf.float32, name='y')
y_val = tf.constant(y_val, dtype=tf.float32, name='y_val')
y_test = tf.constant(y_test, dtype=tf.float32, name='y_test')
```

```python
x_train_shape = x_train.get_shape().as_list()
m = x_train_shape[0]
n = x_train_shape[1]
```

```python
theta =  tf.Variable(tf.random_normal([1, n], stddev=0.35), name="theta")
y_predict = tf.matmul(x_train, tf.transpose(theta), name='Predictions')
error = y_predict - y_train
rmse = tf.sqrt(tf.reduce_mean(tf.square(error)), name='rmse')
scale = np.array([0.001,0.01,1, 10])
learning_rate = 0.01
base_loss = tf.reduce_mean(tf.square(error), name='loss')
reg_loss = tf.reduce_sum(tf.square(theta))
loss = tf.add(base_loss,scale[0]/m * reg_loss)
gradients = (1./m * tf.reduce_sum(tf.multiply(x_train, error), 0))
reduced_theta = theta * (1. - learning_rate * ((2. * scale[0])/m))
training_op = tf.assign(theta, tf.subtract(reduced_theta, learning_rate * gradients))
```

```python
iterations = 1500
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```
