```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error
```

```python
def ryan_initweight(topo):
    np.random.seed(1)
    params = {}
    for i in range(1, len(topo)):
        params['W' + str(i)] = 0.01*np.random.randn(topo[i], topo[i-1])
        params['b' + str(i)] = 0.01*np.zeros(shape=(topo[i], 1))
    return params

def ryan_onehot(labels):
    labels = labels.as_matrix()
    temp = []
    for val in labels:
        if val not in temp:
            temp.append(val)
    temp.sort()
    
    result = np.zeros(shape=(len(labels),len(temp)))
    for key, val in enumerate(labels):
        result[key][temp.index(val)] = 1
    return result

def ryan_sigmoid(W,b,A_prev):
    Z, cache1 = ryan_forward(W, b, A_prev)
    A = 1/(1 + np.exp(-1*Z))
    cache = (cache1, Z)
    return A,cache

def ryan_forward(W,b,A):
    Z = np.dot(W, A) + b
    cache = (A,W,b)
    return Z, cache

def ryan_deep_forward(X, params):
    caches = []
    A = X
    for i in range(1, len(params)//2):
        A_prev = A
        A, cache = ryan_sigmoid(params['W' + str(i)], params['b' + str(i)], A_prev)
        caches.append(cache)
    
    lastA, cache = ryan_sigmoid(params['W' + str(len(params) // 2)], params['b' + str(len(params) // 2)], A)
    caches.append(cache)
    return lastA, caches

def ryan_cost_func(lastA, Y):
    m = Y.shape[1]
    cost = (-1/m) * np.sum(np.multiply(Y, np.log(lastA)) + np.multiply(1 - Y, np.log(1 - lastA)))
    return cost

def ryan_backward(dZ, cache):
    A_prev = cache[0]
    W = cache[1]
    b = cache[2]

    m = A_prev.shape[1]
    
    dW = np.dot(dZ, A_prev.T)/m
    db = np.sum(dZ, axis=1, keepdims=True)/m
    dA_prev = np.dot(W.T, dZ)
    
    return dA_prev, dW, db
    
def ryan_sigmoid_back(dA, cache): #z
    temp = 1/(1+np.exp(-1*cache))
    dZ = np.multiply(dA, np.multiply(temp, (np.subtract(np.ones(temp.shape), temp))))
    return dZ

def ryan_deep_backward(lastA, Y, caches):

    grads = {}
    L = len(caches) # the number of layers
    m = lastA.shape[1]
    
    # Initializing the backpropagation
    dlastA = - (np.divide(Y, lastA) - np.divide(1 - Y, 1 - lastA))
    current_cache = caches[-1]

    grads["dA" + str(L)], grads["dW" + str(L)], grads["db" + str(L)] = ryan_backward(ryan_sigmoid_back(dlastA, current_cache[1]), current_cache[0])
    tt = grads["dA" + str(L)]
    for l in reversed(range(L-1)):
        current_cache = caches[l]
        dA_prev_temp, dW_temp, db_temp = ryan_backward(ryan_sigmoid_back(tt, current_cache[1]), current_cache[0])
        tt = dA_prev_temp
        grads["dA" + str(l + 1)] = dA_prev_temp
        grads["dW" + str(l + 1)] = dW_temp
        grads["db" + str(l + 1)] = db_temp

    return grads

def ryan_update_parameters(params, grads, learning_rate):
    for l in range(len(params) // 2):
        params["W" + str(l + 1)] = params["W" + str(l + 1)] - learning_rate * grads["dW" + str(l + 1)]
        params["b" + str(l + 1)] = params["b" + str(l + 1)] - learning_rate * grads["db" + str(l + 1)]
        
    return params
```

```python
df_train = pd.read_csv('ex3_train.csv.xls');
df_test = pd.read_csv('ex3_test.csv.xls');
```

```python
X_train = df_train.iloc[:,0:400].T
Y_train = ryan_onehot(df_train['y']).T

X_test = df_test.iloc[:,0:400].T
Y_test = ryan_onehot(df_test['y']).T
```

```python
def ryan_gd(X, Y, topo, learning_rate=0.001, iterations=3000):

    params = ryan_initweight(topo)
    
    costs = []
    for i in range(0, iterations):
        lastA, caches = ryan_deep_forward(X, params)
        cost = ryan_cost_func(lastA, Y)
        grads = ryan_deep_backward(lastA, Y, caches)
        params = ryan_update_parameters(params, grads, learning_rate)
                
        # Print the cost every 100 training example
        if i % 100 == 0:
            print ("Cost after iteration %i: %f" % (i, cost))
        if i % 100 == 0:
            costs.append(cost)
            
    # plot the cost
    plt.plot(costs)
    plt.ylabel('cost')
    plt.xlabel('iterations (per tens)')
    plt.title("Learning rate =" + str(learning_rate))
    plt.show()
    
    return params
```

```python
topo = [400, 25, 10]
learning_rate = 0.05
iterations = 10000
optimize = ryan_gd(X_train, Y_train, topo, learning_rate, iterations)
```
